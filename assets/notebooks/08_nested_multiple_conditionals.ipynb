### 4.2. Ticket Validator (Nested Logic) (Exercise 10)
Write a program that:
1. Ask the user for their `ticket_type` ("VIP" or "Standard") and their `bag_count` (int).
2. Print "Access Granted" if the type is "VIP" and `bag_count <= 2`.
3. Print "Bag Fee" if the type is "Standard" and `bag_count > 1`.
4. In any other case, print "Check Info".

````python
# Write your code here
# 1. Ask for ticket_type (string)
# 2. Ask for bag_count (int)
# 3. Use if/elif/else with nested or compound conditions
# 4. Print the result
````

<details><summary><strong>Solution</strong></summary>

```python
ticket_type = input("Enter ticket type (VIP/Standard): ")
bag_count = int(input("Enter bag count: "))
if ticket_type == "VIP":
    if bag_count <= 2:
        print("Access Granted")
    else:
        print("Check Info")
elif ticket_type == "Standard":
    if bag_count > 1:
        print("Bag Fee")
    else:
        print("Access Granted")
else:
    print("Check Info")
```
</details>
