## 4. Mini-Challenge (Exercises 9–10 of 10)

### 4.1. Shopping discount calculator
Ask for the price and the discount percent, compute the final price after discount, and print it.


````python
# Write your code here
# 1. Ask the user for the price and save it as a float
# 2. Ask the user for the discount percent and save it as a float
# 3. Compute the final price after discount
# 4. Print the final price
````

<details><summary><strong>Solution</strong></summary>

```python
price = float(input("Enter price: "))
discount_percent = float(input("Enter discount percent: "))
final_price = price * (1 - discount_percent / 100)
print(final_price)
```
</details>

### 4.2. Body mass index (BMI) calculator
Ask the user for weight in kilograms and height in meters, compute BMI = weight / (height ** 2), and print it.


````python
# Write your code here
# 1. Ask the user for weight in kg and save as float
# 2. Ask the user for height in meters and save as float
# 3. Compute BMI
# 4. Print the BMI
````

<details><summary><strong>Solution</strong></summary>

```python
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))
bmi = weight / (height ** 2)
print(bmi)
```
</details>
