# For Loop

In this notebook, you will practice using the for loop, a structure that repeats code a specific number of times or over a collection of items.

The exercises help you practice:
- Loop structure and iteration
- Using `range()`

You will complete ten exercises to reinforce these concepts.

## 1. Warm-Up Exercise (1 of 10)

### 1.1. Simple Repetition (Exercise 1)
Use a `for` loop to print the message `"Hello"` three times.


````python
# Replace here:
for i in ___:  # <-- Replace with: range(3)
    print("Hello")
````

<details><summary><strong>Solution</strong></summary>


## 4. Mini-Challenge (Exercises 9-10 of 10)

### 4.1. Word Repeater (Exercise 9)
Ask the user for a word and a number `n`. Print the word n times using a `for` loop.


````python
# Write your code here:
# 1. Ask for a word
# 2. Ask for a number
# 3. Repeat the word using a loop
````

<details><summary><strong>Solution</strong></summary>

```python

n = int(input("How many times?: "))
for i in range(n):
    print(word)
```
</details>

### 4.2. Average Calculator (Exercise 10)
Ask the user for five numbers and compute their average.


````python
# Write your code here:
# 1. Initialize total
# 2. Loop 5 times asking for a number
# 3. Compute the average
````

<details><summary><strong>Solution</strong></summary>

```python

    num = float(input("Enter a number: "))

```
