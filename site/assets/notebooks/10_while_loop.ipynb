## 1. Warm-Up Exercise (1 of 10)

### 1.1. Simple Countdown (Exercise 1)
Define a variable `i = 3`. Use a `while` loop to print the numbers 3, 2, 1.

````python
i = 3
while ___:  # <-- Replace with: i > 0
    print(i)
    i = ___  # <-- Replace with: i - 1
````

<details><summary><strong>Solution</strong></summary>

### 4.2. Factorial Calculator (Exercise 10)
Ask the user for a number and compute its factorial using a `while` loop.


````python
# Write your code here:
# 1. Ask for the number
# 2. Use a loop to multiply values
# 3. Print the result
````

<details><summary><strong>Solution</strong></summary>

```python
num = int(input("Enter a number: "))
result = 1
i = 1
while i <= num:
    result = result * i
    i = i + 1
print(result)
```
</details>
